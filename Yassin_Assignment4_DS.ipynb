#### 3. Write a Pandas program to compare the elements of the two Pandas Series.


```python
import pandas as pd
series1 = pd.Series([10, 20, 30, 40, 50])
series2 = pd.Series([10, 25, 30, 35, 50])

equal = series1 == series2
greater = series1 > series2
less = series1 < series2


print("Equal:\n", equal)
print("Greater:\n", greater)
print("Less:\n", less)
```

#### 4. Write a Pandas program to change the data type of given a column or a Series.

```python
import pandas as pd

data = {'A': [1, 2, 3, 4, 5]}
df = pd.DataFrame(data)

df['A'] = df['A'].astype(float)

print(df)
print(df.dtypes)
```

#### 6. Write a Pandas program to display the default index and set a column as an Index in a given dataframe.


```python
import pandas as pd

data = {
    'school': ['Streat1', 'Streat2', 'Streat3', 'Streat4', 'Streat5', 'Streat6'],
    'student_id': ['S001', 'S002', 'S003', 'S004', 'S005', 'S006'],
    'class': ['V', 'V', 'VI', 'VI', 'V', 'VI'],
    'first_name': ['Alberto', 'Gino', 'Ryan', 'Eesha', 'Gino', 'David'],
    'last_name': ['Franco', 'Mcneill', 'Parkes', 'Hinton', 'Mcneill', 'Parkes'],
    'dob': ['15/05/2002', '17/05/2002', '16/02/1999', '25/09/1998', '11/05/2002', '15/09/1997'],
    'age': [35, 32, 33, 30, 31, 32]
}

df = pd.DataFrame(data)


print("Default Index:")
print(df)

df.set_index('student_id', inplace=True)

print("\nDataFrame with 'student_id' as index:")
print(df)
```

#### 7. Write a Pandas program to display the default index and set a column as an Index in a given dataframe and then reset the index


```python
import pandas as pd

data = {
    'school': ['Streat1', 'Streat2', 'Streat3', 'Streat4', 'Streat5', 'Streat6'],
    'student_id': ['S001', 'S002', 'S003', 'S004', 'S005', 'S006'],
    'class': ['V', 'V', 'VI', 'VI', 'V', 'VI'],
    'first_name': ['Alberto', 'Gino', 'Ryan', 'Eesha', 'Gino', 'David'],
    'last_name': ['Franco', 'Mcneill', 'Parkes', 'Hinton', 'Mcneill', 'Parkes'],
    'dob': ['15/05/2002', '17/05/2002', '16/02/1999', '25/09/1998', '11/05/2002', '15/09/1997'],
    'age': [35, 32, 33, 30, 31, 32]
}

df = pd.DataFrame(data)

print("Default Index:")
print(df)

df.set_index('student_id', inplace=True)

print("\nDataFrame with 'student_id' as index:")
print(df)

df.reset_index(inplace=True)

print("\nDataFrame after resetting the index:")
print(df)
```

#### 8. Write a Pandas program to create a dataframe indexing by date and time.

```python
import pandas as pd
import numpy as np


date_range = pd.date_range(start='2024-07-10', end='2024-10-10', freq='H')


data = np.random.rand(len(date_range), 4)


df = pd.DataFrame(data, index=date_range, columns=['A', 'B', 'C', 'D'])

print(df)
```

#### 9. Write a Pandas program to create a dataframe and set a title or name of the index column.


```python
import pandas as pd
import numpy as np

date_range = pd.date_range(start='2023-01-01', end='2023-01-10', freq='H')

data = np.random.rand(len(date_range), 4)

df = pd.DataFrame(data, index=date_range, columns=['A', 'B', 'C', 'D'])

df.index.name = 'DateTime'

print(df)
```
